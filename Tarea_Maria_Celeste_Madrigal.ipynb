```python
#Instalar paquetes
import pandas as pd 
import streamlit as st
import numpy as np
from sklearn import datasets
import pickle
import altair as alt

#El código que contenga las instrucciones para el despliegue de un título y una breve descripción de la aplicación web.
st.title('Análisis del desempeño de los colaboradores de Socialize your Knowledge')
st.text('Esta pagina web tiene como fin análisis de desempeño de los colaboradores de Socialize your Knowledge')

#El código que permita desplegar el logotipo de la empresa en la aplicación web.
from PIL import Image
image = Image.open('/Users/celeste.solorzano/Desktop/Tecnico de Monterrey/Data Scientist /Aplicación web de ciencia de datos/Tarea/imagen.jpeg')
st.image(image, caption='Socialize your Knowledge')
#Importar el csv 

df = pd.read_csv('/Users/celeste.solorzano/Desktop/Tecnico de Monterrey/Data Scientist /Aplicación web de ciencia de datos/0.jfvfpm3cp2g0.58xa4ttfucmemployee_data.csv')
print(df)
st.table (df)
df.info
# El código que permita desplegar un control para seleccionar el género del empleado.
selected_gender = st.selectbox("Seleccione el género", 
df['gender'].unique())
st.write(f"Selected Option: {selected_gender!r}")

# El código que permita desplegar un control para seleccionar un rango del puntaje de desempeño del empleado.
optional = st.beta_expander ("Rango del puntaje de desempeño del empleado", True)
fare_select = optional.slider(
    "Selecione el rango",
    min_value = float(df['performance_score'].min()),
    max_value = float(df['performance_score'].max())
)
subset_fare = df[(df['performance_score'] 
>= fare_select)]
st.write(f"Puntaje de desempeño {fare_select} : {subset_fare.shape[0]}")

# El código que permita desplegar un control para seleccionar el estado civil del empleado.

selected_status = st.selectbox("Seleccione el estado civil", 
df['marital_status'].unique())
st.write(f"Selected Option: {selected_status!r}")

# El código que permita mostrar un gráfico en donde se visualice la distribución de los puntajes de desempeño.
alt.Chart(df).mark_point(filled=True).encode(
    alt.X('performance_score'),
    alt.Y('position'),
    alt.Size('performance_score')
)

#El código que permita mostrar un gráfico en donde se visualice el promedio de horas trabajadas por el género del empleado.
alt.Chart(df).mark_point(filled=True).encode(
    alt.X('average_work_hours'),
    alt.Y('sex'),
    alt.Size('sex')
)

# El código que permita mostrar un gráfico en donde se visualice la edad de los empleados con respecto al salario de estosalt.Chart(df).mark_point(filled=True).encode(
alt.Chart(df).mark_point(filled=True).encode(   
    alt.X('age'),
    alt.Y('salary'),
    alt.Size('salary')
)

# El código que permita mostrar un gráfico en donde se visualice la relación del promedio de horas trabajadas versus el puntaje de desempeño.
alt.Chart(df).mark_point(filled=True).encode(
    alt.X('average_work_hours'),
    alt.Y('performance_score'),
    alt.Size('performance_score')
)

#Conclusión

st.text('Con esta investigación se se puede puede observar un pequeño analisis del desempeño de los empleados de Socialize your Knowledge')
```
